```python
import pandas as pd
import numpy as np
```

```python
schools = "raw_data/schools_complete.csv"
students = "raw_data/students_complete.csv"

schools_pd = pd.read_csv(schools)
students_pd = pd.read_csv(students)
schools_pd.head()
```

```python
students_pd.head()
```

```python
schools_pd = schools_pd.rename(columns={'name':'school'})
students_pd = students_pd.rename(columns={'name':'student'})
schools_cnt = schools_pd["school"].count()
student_tot = students_pd["student"].count()
budget_tot = schools_pd['budget'].sum()
avg_math = students_pd['math_score'].mean()
tot_passM = len(students_pd[students_pd["math_score"]>=70])
avg_reading = students_pd['reading_score'].mean()
tot_passR = len(students_pd[students_pd["reading_score"]>=70])
per_passM = tot_passM/student_tot
per_passR = tot_passR/student_tot
ovr_pass = (per_passM+per_passR)/2

sum_df = pd.DataFrame({
    "Total Schools":[schools_cnt],
    "Total Students":[student_tot],
    "Total Budget":[budget_tot],
    "Average Math Score":[avg_math],
    "Average Reading Score":[avg_reading],
    "Percent Pass Math":[per_passM],
    "PercentPassing Reading":[per_passR],
    "Overall Pass Rate":[ovr_pass]
})
sum_df
```

```python
schools_pd = schools_pd.rename(columns={'name':'school'})
student_grp = students_pd.groupby(['school'])
student_cnt = student_grp["Student ID"].count()
grp_passM = students_pd[students_pd["math_score"]>=70].groupby(["school"])
grp_passM = grp_passM['math_score'].count()
grp_passR = students_pd[students_pd["reading_score"]>=70].groupby(["school"])
grp_passR = grp_passR['reading_score'].count()
avg_passM =  students_grp['math_score'].mean()
avg_passR =  students_grp['reading_score'].mean()

schoolSum = pd.DataFrame({'Students':student_cnt,'Number Passing Math':grp_passM, 'Avg Math Score':avg_passM, 'Number Passing Reading':grp_passR, 'Avg Reading Score':avg_passR})
schoolSum = schoolSum.reset_index()
schoolSum.columns

merge = pd.merge(schoolSum,schools_pd[["school",'type','budget']],on='school')
merge['Budget Per Student']= merge['budget']/merge['Students']
merge['% Passing Math']= round((merge['Number Passing Math'] / merge['Students'])*100,2)
merge['% Passing Reading']= round((merge['Number Passing Reading'] / merge['Students'])*100,2)
merge['Overall Passing']= round((merge['% Passing Math'] + merge['% Passing Reading'])/2,2)
merge.head()
```

```python
ovr_df = merge.sort_values("Overall Passing", ascending=False)
ovr_df.head()
```

```python
ovr_df = merge.sort_values("Overall Passing", ascending=True)
ovr_df.head()
```

```python
bins = [0,1000000,2000000,3000000,10000000]
group_labels = ["0 to 1mil","1mil to 2mil","2mil to 3mil","3mil to 10mil"]
pd.cut(merge["budget"],bins,labels=group_labels).head()
merge["Budget Group"] = pd.cut(merge["budget"],bins,labels=group_labels)
merge.head()
```

```python
# Create a GroupBy object based upon "View Group"
merge_group = merge.groupby(["Budget Group"])
merge_M = merge_group['Avg Math Score'].mean()
merge_R = merge_group['Avg Reading Score'].mean()
per_M = merge_group['% Passing Math'].mean()
per_R = merge_group['% Passing Reading'].mean()
per_over = (per_M+per_R)/2

group_df = pd.DataFrame({'Avg Math Score':merge_M, 'Avg Reading Score':merge_R,'% Passing Math':per_M,'% Passing Reading':per_R,'Overall Pass %':per_over})
group_df
```

```python
bins = [0,1000,2000,10000]
group_labels = ["0 to 1k (Small)","1k to 2k (Medium)","2k to 10k (Large)"]
pd.cut(merge["Students"],bins,labels=group_labels).head()
merge["School Size"] = pd.cut(merge["Students"],bins,labels=group_labels)
merge.head()
```

```python
merge_group = merge.groupby(["School Size"])
merge_M = merge_group['Avg Math Score'].mean()
merge_R = merge_group['Avg Reading Score'].mean()
per_M = merge_group['% Passing Math'].mean()
per_R = merge_group['% Passing Reading'].mean()
per_over = (per_M+per_R)/2

group_df = pd.DataFrame({'Avg Math Score':merge_M, 'Avg Reading Score':merge_R,'% Passing Math':per_M,'% Passing Reading':per_R,'Overall Pass %':per_over})
group_df
```
